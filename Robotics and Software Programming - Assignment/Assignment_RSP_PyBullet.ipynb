```python
import pybullet as p
import pybullet_data
import matplotlib.pyplot as plt
import numpy as np

# 1. Connect to the physics server in DIRECT mode
# This mode runs the simulation without a graphical window, which is ideal for
# scripting and capturing images.
try:
    p.disconnect()
except:
    pass
physicsClient = p.connect(p.DIRECT)
p.setAdditionalSearchPath(pybullet_data.getDataPath())
p.setGravity(0, 0, -9.81)

# 2. Load the ground plane
planeId = p.loadURDF("plane.urdf")

# 3. Define object positions and orientations
# A dictionary is used to manage the scene configuration cleanly.
# At least seven objects plus the robot must be instantiated.
scene_objects = {
    "wall_back": {"path": "wall.urdf", "pos": [0, 3, 0.5], "orn": p.getQuaternionFromEuler([0, 0, 1.57])},
    "wall_left": {"path": "wall.urdf", "pos": [-1.5, 1.5, 0.5], "orn": p.getQuaternionFromEuler([0, 0, 1.57])},
    "table_1": {"path": "box_table.urdf", "pos": [1, 1, 0.2], "orn": p.getQuaternionFromEuler([0, 0, 0.5])},
    "table_2": {"path": "box_table.urdf", "pos": [-1, -1.5, 0.2], "orn": p.getQuaternionFromEuler([0, 0, -0.8])},
    "pillar_1": {"path": "cylinder_pillar.urdf", "pos": [2, -2, 0.75], "orn": p.getQuaternionFromEuler([0, 0, 1.57])},
    "pillar_2": {"path": "cylinder_pillar.urdf", "pos": [-2, 2, 0.75], "orn": p.getQuaternionFromEuler([0, 0, 1.57])},
    "obstacle_1": {"path": "sphere_obstacle.urdf", "pos": [0, -1, 0.25], "orn": p.getQuaternionFromEuler([0, 0, 1.57])}
}

# Load static objects into the simulation
for key, obj in scene_objects.items():
    p.loadURDF(fileName=obj["path"], basePosition=obj["pos"], baseOrientation=obj["orn"])

# 4. Define robot positions for the three images
robot_poses = [
    {"pos": [-1, 0, 0.1], "orn": p.getQuaternionFromEuler([0, 0, 0.2])},
    {"pos": [0, 0.5, 0.1], "orn": p.getQuaternionFromEuler([0, 0, 1.0])},
    {"pos": [1, -0.5, 0.1], "orn": p.getQuaternionFromEuler([0, 0, -0.5])}
]

```

```python
import random
def get_random_view():
    return (random.randint(-10, 10), random.randint(-10, 10), random.randint(-10, 10))
```

```python
def capture_image(robot_pose, camera_config):
    """Loads the robot at a specific pose and captures an image."""
    robotId = p.loadURDF("two_wheel_robot.urdf", robot_pose["pos"], robot_pose["orn"])

    # Camera parameters
    width, height = 640, 480
    fov, aspect, near, far = 60, width / height, 0.1, 100

    # Projection matrix defines the camera's lens properties.
    projection_matrix = p.computeProjectionMatrixFOV(fov, aspect, near, far)

    # View matrix defines the camera's position and target.
    view_matrix = p.computeViewMatrix(
        cameraEyePosition=camera_config["eye_pos"],
        cameraTargetPosition=camera_config["target_pos"],
        cameraUpVector=camera_config["up_vector"]
    )

    # Get the image data from PyBullet.
    width, height, img_data, depthImg, segImg = p.getCameraImage(width, height, view_matrix, projection_matrix, renderer=p.ER_BULLET_HARDWARE_OPENGL)

    # PyBullet returns a tuple; the third element is the RGB image array.
    rgb_image = np.reshape(img_data, (height, width, 4))[:, :, :3] # Drop alpha channel

    p.removeBody(robotId) # Clean up robot for next capture
    return rgb_image

# Define the three camera configurations
camera_configs = {
    "top_down": {
        "eye_pos": get_random_view(),
        "target_pos": get_random_view(),
        "up_vector": get_random_view()
    },
    "isometric": {
        "eye_pos": get_random_view(),
        "target_pos": get_random_view(),
        "up_vector": get_random_view()
    },
    "side_view": {
        "eye_pos": [-5, 0, 1],
        "target_pos": [0, 0, 0.5],
        "up_vector": get_random_view()
    }
}

# Capture images from each viewpoint. The robot moves between each shot.
images = []
camera_viewpoints = list(camera_configs.values())
print(camera_viewpoints)

for i in range(3):
    img = capture_image(robot_poses[i], camera_viewpoints[i])
    images.append(img)

# Display the captured images using Matplotlib
fig, axs = plt.subplots(1, 3, figsize=(18, 6))
titles =[]
for i, ax in enumerate(axs):
    ax.imshow(images[i])
    # ax.set_title(titles[i])
    ax.axis('off')

plt.tight_layout()
plt.show()

```

```python
import pybullet as p
import pybullet_data
import os
import random
import matplotlib.pyplot as plt

class Scene:
    def __init__(self):
        self.objects = []

    def build_warehouse(self):
        """Create ground + surrounding walls + some obstacles"""
        p.setAdditionalSearchPath(pybullet_data.getDataPath())
        plane = p.loadURDF("plane.urdf")
        self.objects.append(plane)

        # Example: 4 walls around (like a box 5x5)
        wall_thickness = 0.1
        wall_height = 1.0
        size = 15

        texture_id = p.loadTexture("brick.png")  # place brick.png in same dir
        wall_urdf = "wall.urdf"  # You must define a simple wall URDF
        self.objects.append(
            p.loadURDF(wall_urdf, [size/2, 0, 0.5], p.getQuaternionFromEuler([0,0,1.57]), useFixedBase=True)
        )
        self.objects.append(
            p.loadURDF(wall_urdf, [-size/2, 0, 0.5], p.getQuaternionFromEuler([0,0,1.57]), useFixedBase=True)
        )

        wall1 = p.loadURDF(wall_urdf, [0, size/2, 0.5], useFixedBase=True)
        p.changeVisualShape(wall1, -1, textureUniqueId=texture_id)
        self.objects.append(wall1)
        self.objects.append(p.loadURDF(wall_urdf, [0,-size/2, 0.5], useFixedBase=True))


        # Add some extra fixed objects
        self.objects.append(p.loadURDF("box_object.urdf", [1, 1, 0.2]))
        self.objects.append(p.loadURDF("cylinder_pillar.urdf", [-1, -1, 0.2]))
        return self.objects


class Camera:
    def __init__(self, target=[0,0,0], distance=5):
        self.target = target
        self.distance = distance
        self.view_matrix = None
        self.proj_matrix = p.computeProjectionMatrixFOV(
            fov=90, aspect=1.0, nearVal=0.1, farVal=20
        )

    def set_view(self, view="top", yaw=None, pitch=None, roll=0):
        """Set camera view by predefined name or custom yaw/pitch"""
        if view == "top":
            yaw, pitch = 0, -89
        elif view == "left":
            yaw, pitch = 90, -30
        elif view == "right":
            yaw, pitch = -90, -30
        elif view == "rear":
            yaw, pitch = 180, -30


        self.view_matrix = p.computeViewMatrixFromYawPitchRoll(
            cameraTargetPosition=self.target,
            distance=self.distance,
            yaw=yaw,
            pitch=pitch,
            roll=roll,
            upAxisIndex=2
        )
        return self.view_matrix

    def capture(self, filename=None):
        """Capture current view and save image"""
        width, height, rgbImg, depthImg, segImg = p.getCameraImage(
            width=1920, height=1280,
            viewMatrix=self.view_matrix,
            projectionMatrix=self.proj_matrix
        )
        rgbImg = np.reshape(rgbImg, (height, width, 4)).astype(np.uint8)
        img = rgbImg[:, :, :3]

        if filename is None:
            filename = "out/capture.png"
        os.makedirs(os.path.dirname(filename), exist_ok=True)
        plt.imsave(filename, img)
        print(f"Saved: {filename}")



class Robot:
    def __init__(self, urdf_path="two_wheel_robot.urdf"):
        p.connect(p.DIRECT)
        p.setGravity(0,0,-9.8)

        # load scene
        self.scene = Scene()
        self.scene.build_warehouse()

        # load robot
        self.robot_id = p.loadURDF(urdf_path, [0,0,0.1])

        # attach camera
        self.camera = Camera(
            target=[0,0,0], distance=10
        )

    def move_to(self, pos):
        """Teleport robot to given pos"""
        p.resetBasePositionAndOrientation(self.robot_id, pos, [0,0,0,1])

    def teleport(self, bounds=[-2,2]):
        """Teleport robot to random location inside given bounds"""
        x = random.uniform(bounds[0], bounds[1])
        y = random.uniform(bounds[0], bounds[1])
        self.move_to([x, y, 0.1])

    def disconnect(self):
        p.disconnect()
```

```python
import pybullet as p
import pybullet_data
import os
import random
import matplotlib.pyplot as plt

class Scene:
    def __init__(self, start_x=0, start_y=0):
        self.objects = []
        self.start_x, self.start_y = start_x, start_y

    def build_platforms(self):
        """Create ground + surrounding walls + some obstacles"""
        p.setAdditionalSearchPath(pybullet_data.getDataPath())
        plane = p.loadURDF("plane.urdf")
        texture_id = p.loadTexture("brick.png") 
        self.objects.append(plane)
        
        platform_1 = p.loadURDF("platform_1.urdf", [self.start_x, self.start_y, 0.0], p.getQuaternionFromEuler([0,0,1.57]), useFixedBase=True)
        p.changeVisualShape(platform_1, -1, textureUniqueId=texture_id)
        platform_2 = p.loadURDF("platform_2.urdf", [self.start_x + 2, self.start_y, 0.0], p.getQuaternionFromEuler([0,0,1.57]), useFixedBase=True)
        p.changeVisualShape(platform_2, -1, textureUniqueId=texture_id)

        platform_3 = p.loadURDF("platform_3.urdf", [self.start_x + 2, self.start_y + 2, 0.0], p.getQuaternionFromEuler([0,0,1.57]), useFixedBase=True)
        platform_4 = p.loadURDF("platform_4.urdf", [self.start_x , self.start_y+2, 0.0], p.getQuaternionFromEuler([0,0,1.57]), useFixedBase=True)

        for plfm in [platform_1, platform_2, platform_3, platform_4]:
            self.objects.append(plfm)
        return self.objects


class Camera:
    def __init__(self, target=[0,0,0], distance=5, fov=60):
        self.target = target
        self.distance = distance
        self.view_matrix = None
        self.proj_matrix = p.computeProjectionMatrixFOV(
            fov=fov, aspect=1.0, nearVal=0.1, farVal=20
        )
        
    def set_distance(self, distance: int):
        self.distance = distance

    def set_view(self, view="top", yaw=None, pitch=None, roll=0):
        """Set camera view by predefined name or custom yaw/pitch"""
        if view == "top":
            yaw, pitch = 0, -89
        elif view == "left":
            yaw, pitch = 90, -30
        elif view == "right":
            yaw, pitch = -90, -30
        elif view == "rear":
            yaw, pitch = 180, -30
        else:
            if yaw is None and pitch is None:
                print('If view is none, you should pass yaw and pitch. otherwise, shows default values!')
                yaw, pitch = 0, -89

        self.view_matrix = p.computeViewMatrixFromYawPitchRoll(
            cameraTargetPosition=self.target,
            distance=self.distance,
            yaw=yaw,
            pitch=pitch,
            roll=roll,
            upAxisIndex=2
        )
        return self.view_matrix

    def capture(self, filename=None):
        """Capture current view and save image"""
        width, height, rgbImg, depthImg, segImg = p.getCameraImage(
            width=640, height=480,
            viewMatrix=self.view_matrix,
            projectionMatrix=self.proj_matrix
        )
        rgbImg = np.reshape(rgbImg, (height, width, 4)).astype(np.uint8)
        img = rgbImg[:, :, :3]

        if filename is None:
            filename = "out/capture.png"
        os.makedirs(os.path.dirname(filename), exist_ok=True)
        plt.imsave(filename, img)
        print(f"Saved: {filename}")



class Robot:
    def __init__(self, urdf_path="two_wheel_robot.urdf"):
        p.connect(p.DIRECT)
        p.setGravity(0,0,-9.8)
        self.start_x, self.start_y, self.start_z = -1, -1, 2

        # load scene
        self.scene = Scene(start_x, start_y)
        self.scene.build_platforms()

        # load robot
        self.robot_id = p.loadURDF(
            urdf_path, 
            [self.start_x, self.start_y, self.start_z]
        )

        # attach camera
        self.camera = Camera(
            target=[0,0,0], distance=10
        )

    def move_to(self, pos, quaternion=[0, 0, 0, 1]):
        """Teleport robot to given pos"""
        p.resetBasePositionAndOrientation(self.robot_id, pos, quaternion)

    def teleport(self, bounds=[-2,2]):
        """Teleport robot to random location inside given bounds"""
        x = random.uniform(bounds[0], bounds[1])
        y = random.uniform(bounds[0], bounds[1])
        self.move_to([x, y, self.start_z])

    def disconnect(self):
        p.disconnect()
```

```python
import math
sim = Robot()
# Move robot to 3 positions, capture from top/left/right
positions = [[-1,-1, 2], [1, 1, 2], [1,-1, 2], [-1, 1, 2.5]]
views = ["top", "left", "rear"]

for i, pos in enumerate(positions):
    if i == 3:
        yaw_angle_rad_3 = -90 * (math.pi / 180) # Negative angle for right turn
        orientation3 = p.getQuaternionFromEuler([0, 0, yaw_angle_rad_3])
        sim.move_to(pos, orientation3)
        sim.camera.set_view(view=views[2])
        sim.camera.capture(f"out/scene_{i+1}.png")
    else:
        sim.move_to(pos)
        sim.camera.set_view(view=views[2])
        sim.camera.capture(f"out/scene_{i+1}.png")

from PIL import Image
from IPython.display import display

outf = 'out'
imgs = [os.path.join(outf, i) for i in os.listdir(outf) if i.endswith(('jpg', 'png'))]

for i in imgs:
    display(Image.open(i))
    
sim.disconnect()
```
